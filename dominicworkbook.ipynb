```python
import re
from pathlib import Path
from src.database_read.ED_Visits import extract_sheps_ed_2021_final1

pdf_path = Path("data/ED_visits/ed_ptchar_all_and_by_hosp_2021_and.pdf")
df_wide = extract_sheps_ed_2021_final1(pdf_path)
df_wide.head(50)
```

```python
df_wide.head(50)
```

```python
import pandas as pd

df = df_wide.copy()

df.columns = ["County", "Hospital", "Section", "Metric", "N_2021", "Pct_2021"]

# Clean Section and Metric text
df["Section"] = df["Section"].str.replace(r"^N\s*%\s*N.*?Patient", "Patient", regex=True)
df["Section"] = df["Section"].str.strip()
df["Metric"] = df["Metric"].str.strip()

# Fix data types
df["N_2021"] = pd.to_numeric(df["N_2021"], errors="coerce")
df["Pct_2021"] = pd.to_numeric(df["Pct_2021"], errors="coerce")

# Drop summary rows (None counties)
df = df[df["County"].notna() & (df["County"].str.lower() != "none")]

# Sort nicely
df = df.sort_values(["County", "Hospital", "Section"]).reset_index(drop=True)

print(df.shape)
print(df.head(10))
```
