```python
import pandas as pd
from pathlib import Path

# Path to your directory
data_dir = Path("data/MA/2023_fy")

# Correct filename pattern
files = sorted(data_dir.glob("SCC_Enrollment_MA_2022_*.csv")) + \
        sorted(data_dir.glob("SCC_Enrollment_MA_2023_0[1-6].csv"))

if not files:
    raise FileNotFoundError(f"No enrollment files found in {data_dir}")

print(f"Found {len(files)} monthly files.")

all_months = []

for file in files:
    # Extract month from filename (e.g., 2022_07)
    month = "_".join(file.stem.split("_")[-2:])

    # Read safely with encoding handling
    df = pd.read_csv(file, dtype=str, encoding="latin1", on_bad_lines="skip")
    df.columns = [c.strip().lower() for c in df.columns]

    # Confirm expected columns exist
    if not {"state", "county", "enrolled"}.issubset(df.columns):
        print(f" Skipping {file.name}: Missing expected columns")
        continue

    # Filter for NC and ensure numeric enrollment
    df_nc = df[df["state"].str.upper() == "NC"].copy()
    df_nc["enrolled"] = pd.to_numeric(df_nc["enrolled"], errors="coerce").fillna(0)

    # Aggregate by county
    county_sum = df_nc.groupby("county", as_index=False)["enrolled"].sum()
    county_sum.rename(columns={"enrolled": month}, inplace=True)
    all_months.append(county_sum)

# Merge all monthly county totals
df_final = all_months[0]
for df_month in all_months[1:]:
    df_final = pd.merge(df_final, df_month, on="county", how="outer")

# Clean up
df_final = df_final.fillna(0)
for col in df_final.columns[1:]:
    df_final[col] = df_final[col].astype(int)

# Optional: add total enrollments across all months
df_final["Total_Enrollments"] = df_final.iloc[:, 1:].sum(axis=1)

# Save
output_path = Path("data/final/MA_enrollees_NC_2022_07_to_2023_06.csv")
output_path.parent.mkdir(parents=True, exist_ok=True)
df_final.to_csv(output_path, index=False)

print(f" Saved to {output_path}")
print(df_final.head())
```

```python
import pandas as pd
from pathlib import Path

# --- Load Data ---
ed_path = Path("data/final_bad/sheps_ed_2021_wide_final1.csv")
counties_path = Path("data/County List/county_names.csv")

df_ed = pd.read_csv(ed_path, dtype=str)

# County list has no header, so name manually
df_counties = pd.read_csv(counties_path, header=None, names=["COUNTY"], dtype=str)
df_counties["COUNTY"] = df_counties["COUNTY"].str.strip().str.upper()
county_col = "COUNTY"

# --- Clean and Convert ---
df_ed["County"] = df_ed["County"].str.strip().str.upper()
df_ed["N_2021"] = pd.to_numeric(df_ed["N_2021"], errors="coerce").fillna(0)

# --- Aggregate by County, Section, Metric ---
agg = (
    df_ed
    .groupby(["County", "Section", "Metric"], as_index=False)["N_2021"]
    .sum()
)

# --- Pivot wide ---
pivot = (
    agg.pivot_table(
        index="County",
        columns=["Section", "Metric"],
        values="N_2021",
        fill_value=0
    )
)

# Flatten MultiIndex columns (Section_Metric)
pivot.columns = [f"{sec}_{met}".replace(" ", "_") for sec, met in pivot.columns]
pivot = pivot.reset_index()

# --- Merge with full county list to include missing ones ---
final_df = df_counties.merge(
    pivot, left_on=county_col, right_on="County", how="left"
).drop(columns=["County"])

# Replace missing hospital data with zeros
for col in final_df.columns[1:]:
    final_df[col] = pd.to_numeric(final_df[col], errors="coerce").fillna(0)

# --- Save cleaned, aggregated version ---
out_path = Path("data/final/ed_visits_by_county.csv")
final_df.to_csv(out_path, index=False)

print(f" Saved aggregated ED visits by county to: {out_path}")
print(f"Shape: {final_df.shape}")
print("\nPreview:")
print(final_df.head())
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import r2_score, mean_squared_error

# -----------------------
# Config / Paths
# -----------------------
DATA_PATH = Path("data/final/merged_county_data_removed_dups.csv")
OUT_DIR   = Path("data/final/model")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# -----------------------
# Helpers
# -----------------------
def find_total_ed_column(df: pd.DataFrame) -> str:
    """
    Try to locate/standardize the target column.
    Priority:
      - 'Total_ED'
      - any col containing 'Total_Number_of_Visits' (e.g., 'Payer_Total_Number_of_Visits')
    """
    cols_lower = {c.lower(): c for c in df.columns}
    if "total_ed" in cols_lower:
        return cols_lower["total_ed"]
    candidates = [c for c in df.columns if "total_number_of_visits" in c.lower()]
    if candidates:
        return candidates[0]
    raise ValueError("Could not find a Total ED column. Expected 'Total_ED' or a '*Total_Number_of_Visits' column.")

def numericize(df: pd.DataFrame, exclude_cols=("County", "Region")) -> pd.DataFrame:
    """Coerce all non-excluded columns to numeric (errors='coerce')."""
    out = df.copy()
    for c in out.columns:
        if c not in exclude_cols:
            out[c] = pd.to_numeric(out[c], errors="coerce")
    return out

def quick_eda(df: pd.DataFrame, target_col: str, tag: str):
    """Save EDA snapshots: missingness and top absolute correlations with target."""
    eda_dir = OUT_DIR / f"eda_{tag}"
    eda_dir.mkdir(exist_ok=True, parents=True)

    # Missingness
    miss = df.isna().mean().sort_values(ascending=False).to_frame("missing_pct")
    miss.to_csv(eda_dir / "missingness.csv")

    # Correlations to target (only numeric)
    numdf = df.select_dtypes(include=[np.number])
    if target_col not in numdf.columns:
        # might be all NaN if not numeric; skip correlations
        return
    corrs = numdf.corr(numeric_only=True)[target_col].dropna().sort_values(key=np.abs, ascending=False)
    corrs.to_csv(eda_dir / "correlations_to_target.csv")

from sklearn.linear_model import LinearRegression, RidgeCV, Lasso

def evaluate_models(X, y, tag: str):
    """
    Fit Linear Regression, RidgeCV (L2), and Lasso (L1) without cross-validation.
    Save metrics + coefficients.
    """
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42
    )

    base_steps = [
        ("impute", SimpleImputer(strategy="median")),
        ("scale", StandardScaler(with_mean=True, with_std=True)),
    ]

    results = []

    # 1) Linear
    lr = Pipeline(base_steps + [("model", LinearRegression())])
    lr.fit(X_train, y_train)
    yhat = lr.predict(X_test)
    results.append(("Linear", None, r2_score(y_test, yhat), np.sqrt(mean_squared_error(y_test, yhat))))

    # 2) Ridge (auto alpha, no cv)
    ridge = Pipeline(base_steps + [("model", RidgeCV(alphas=np.logspace(-3, 3, 50)))])
    ridge.fit(X_train, y_train)
    yhat = ridge.predict(X_test)
    ridge_alpha = ridge.named_steps["model"].alpha_
    results.append(("Ridge", ridge_alpha, r2_score(y_test, yhat), np.sqrt(mean_squared_error(y_test, yhat))))

    # 3) Lasso (fixed alpha)
    lasso = Pipeline(base_steps + [("model", Lasso(alpha=0.1, max_iter=10000, random_state=42))])
    lasso.fit(X_train, y_train)
    yhat = lasso.predict(X_test)
    results.append(("Lasso", 0.1, r2_score(y_test, yhat), np.sqrt(mean_squared_error(y_test, yhat))))

    # Save metrics
    metrics_df = pd.DataFrame(results, columns=["Model", "Alpha", "R2", "RMSE"])
    metrics_df.to_csv(OUT_DIR / f"metrics_{tag}.csv", index=False)

    # Coefficients
    coeffs = []
    for model_name, model in [
        ("Linear", lr.named_steps["model"]),
        ("Ridge", ridge.named_steps["model"]),
        ("Lasso", lasso.named_steps["model"]),
    ]:
        try:
            coeffs.append(pd.DataFrame({
                "Feature": X.columns,
                "Coefficient": model.coef_,
                "Model": model_name
            }))
        except Exception as e:
            print(f"[Warning] No coefficients for {model_name}: {e}")

    if coeffs:
        pd.concat(coeffs, ignore_index=True).to_csv(OUT_DIR / f"coefficients_{tag}.csv", index=False)

    return metrics_df


# -----------------------
# Load + Prepare
# -----------------------
df = pd.read_csv(DATA_PATH, dtype=str)
# Standardize County/Region casing
if "County" in df.columns:
    df["County"] = df["County"].astype(str).str.strip().str.title()
elif "COUNTY" in df.columns:
    df.rename(columns={"COUNTY": "County"}, inplace=True)
    df["County"] = df["County"].astype(str).str.strip().str.title()

if "Region" in df.columns:
    df["Region"] = df["Region"].astype(str).str.strip().str.title()
else:
    df["Region"] = "Unknown"

# Numericize features (keep County/Region as is)
df = numericize(df, exclude_cols=("County", "Region"))

# Identify/create target
target_col = None
try:
    target_col = find_total_ed_column(df)
except ValueError:
    # Optional: try to construct a proxy by summing ED metrics if needed (commented out by default)
    # ed_like = [c for c in df.columns if "total_number_of_visits" in c.lower()]
    # if ed_like:
    #     target_col = "Total_ED"
    #     df[target_col] = df[ed_like].sum(axis=1, min_count=1)
    raise

print(f"Using target column: {target_col}")

# -----------------------
# EDA (statewide)
# -----------------------
quick_eda(df, target_col, tag="statewide_raw")

# -----------------------
# Modeling  Statewide (exclude counties with 0 Total_ED)
# -----------------------
df_state = df.copy()
df_state = df_state[df_state[target_col].fillna(0).astype(float) > 0].copy()

feature_cols = [c for c in df_state.columns if c not in ["County", "Region", target_col]]
X_state = df_state[feature_cols].astype(float)
y_state = df_state[target_col].astype(float)

metrics_state = evaluate_models(X_state, y_state, tag="statewide")
print("Statewide metrics saved to:", OUT_DIR / "metrics_statewide.csv")

# -----------------------
# Modeling  Per Region (exclude 0 ED within region)
# -----------------------
regions = sorted(df["Region"].dropna().unique().tolist())
all_region_metrics = []

for r in regions:
    sub = df[df["Region"] == r].copy()
    if sub.empty:
        continue
    sub = sub[sub[target_col].fillna(0).astype(float) > 0]
    if sub.shape[0] < 10:
        # Not enough data to cross-validate robustly; still try with a warning
        print(f" Region '{r}' has only {sub.shape[0]} rows with ED>0. Proceeding, but results may be unstable.")

    if sub.shape[0] < 3:
        print(f"Skipping region '{r}' (too few rows after filtering).")
        continue

    Xr = sub[[c for c in sub.columns if c not in ["County", "Region", target_col]]].astype(float)
    yr = sub[target_col].astype(float)

    tag = f"region_{r.replace(' ', '_')}"
    quick_eda(sub, target_col, tag=tag)
    m = evaluate_models(Xr, yr, tag=tag)
    m.insert(0, "Region", r)
    all_region_metrics.append(m)

if all_region_metrics:
    pd.concat(all_region_metrics, ignore_index=True).to_csv(OUT_DIR / "metrics_by_region.csv", index=False)
    print("Per-region metrics saved to:", OUT_DIR / "metrics_by_region.csv")

print("\n Done. Artifacts written to:", OUT_DIR)
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

IN_PATH  = Path("data/final/merged_county_data_removed_dups.csv")
OUT_DIR  = Path("data/final/need_scores")
OUT_DIR.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(IN_PATH, dtype=str)

# ---- Basic cleaning
for c in ["County", "Region"]:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip().str.title()
    else:
        df[c] = "Unknown"

# Coerce numerics
num_cols = [c for c in df.columns if c not in ["County", "Region"]]
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# ---- Build components

# 1) Public enrollees = Medicaid + Medicare Advantage (proxy for total public coverage)
df["Public_Enrollees"] = df.get("AVG_MC_Enrollees", 0).fillna(0) + df.get("AVG_MA_Enrollments_2023_FY", 0).fillna(0)

# 2) Primary care provider supply (prefer per-cap  headcount; fallback to spec sum)
#    PCP per 100k  absolute count using total population
pcp_percap = df.get("MDs_PrimaryCare_perCap_2021")
pop        = df.get("E_TOTPOP")

pcp_from_percap = pcp_percap * pop / 100000.0
pcp_from_spec   = (
    df.get("MDs_FamilyMed_2021", 0).fillna(0) +
    df.get("MDs_InternalMed_2021", 0).fillna(0) +
    df.get("MDs_Pediatrics_2021", 0).fillna(0)
)

df["PCP_Count"] = np.where(pcp_percap.notna() & pop.notna(), pcp_from_percap, pcp_from_spec)
# Guard against zeros
df["PCP_Count"] = df["PCP_Count"].clip(lower=0)

# 3) Availability Gap
eps = 1e-6
df["Availability_Gap"] = df["Public_Enrollees"] / (df["PCP_Count"] + eps)

# 4) Avoidable Utilization Burden (proxy: ED per 1k public enrollees)
if "Total_ED" not in df.columns:
    # Try to find *Total_Number_of_Visits
    candidates = [c for c in df.columns if "total_number_of_visits" in c.lower()]
    if candidates:
        df["Total_ED"] = df[candidates[0]]
    else:
        raise ValueError("No Total_ED (or *Total_Number_of_Visits) column found in merged file.")
df["Avoidable_Util_Burden"] = df["Total_ED"] / (df["Public_Enrollees"] / 1000.0 + eps)

# 5) Risk Index = SVI overall + mean of available chronic/behavioral prevalences
#    Normalize SVI to 0..1 if it looks like 0..100
svi = df.get("RPL_THEMES")
if svi is None:
    svi_norm = pd.Series(np.nan, index=df.index)
else:
    svi_norm = svi.copy()
    # If max > 1.5 assume it's 0..100 scale and divide by 100
    if svi_norm.max(skipna=True) and (svi_norm.max(skipna=True) > 1.5):
        svi_norm = svi_norm / 100.0

# Chronic/behavioral candidates: include whichever exist
chronic_cols = [
    "OBESITY_age-adjusted_prevalence",
    "DIABETES_age-adjusted_prevalence",
    "CSMOKING_age-adjusted_prevalence",
    "LPA_age-adjusted_prevalence",
    "CHD_age-adjusted_prevalence",
]
available_chronic = [c for c in chronic_cols if c in df.columns]
if not available_chronic:
    # Fall back to crude if age-adjusted missing
    chronic_cols_crude = [c.replace("age-adjusted_", "") for c in chronic_cols]
    available_chronic = [c for c in chronic_cols_crude if c in df.columns]

if available_chronic:
    chronic_pct = df[available_chronic].mean(axis=1, skipna=True) / 100.0  # to 0..1
else:
    chronic_pct = pd.Series(np.nan, index=df.index)

df["Risk_Index_raw"] = svi_norm + chronic_pct

# ---- Scale each metric to 0100 (higher = more need)
def minmax(series):
    s = series.astype(float)
    lo, hi = s.min(skipna=True), s.max(skipna=True)
    if pd.isna(lo) or pd.isna(hi) or hi == lo:
        return pd.Series(np.nan, index=s.index)
    return (s - lo) / (hi - lo) * 100.0

df["Availability_Gap_Score"]   = minmax(df["Availability_Gap"])
df["Util_Burden_Score"]        = minmax(df["Avoidable_Util_Burden"])
df["Risk_Index_Score"]         = minmax(df["Risk_Index_raw"])

# ---- Composite (simple average)
df["Composite_Need_Score"] = df[["Availability_Gap_Score", "Util_Burden_Score", "Risk_Index_Score"]].mean(axis=1)

# ---- Output
keep_cols = [
    "County","Region",
    "Public_Enrollees","PCP_Count","Availability_Gap","Avoidable_Util_Burden","Risk_Index_raw",
    "Availability_Gap_Score","Util_Burden_Score","Risk_Index_Score","Composite_Need_Score",
    "Total_ED","E_TOTPOP","AVG_MC_Enrollees","AVG_MA_Enrollments_2023_FY",
    "RPL_THEMES"
] + available_chronic

out = df[keep_cols].sort_values("Composite_Need_Score", ascending=False)

# Save statewide table + per-region rankings
out.to_csv(OUT_DIR / "county_need_scores_statewide.csv", index=False)

for r, sub in out.groupby("Region", dropna=False):
    fname = f"county_need_scores_region_{str(r).replace(' ','_')}.csv"
    sub.sort_values("Composite_Need_Score", ascending=False).to_csv(OUT_DIR / fname, index=False)

print(" Wrote:")
print(" -", OUT_DIR / "county_need_scores_statewide.csv")
print(" - per-region CSVs in", OUT_DIR)
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import RidgeCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Prepare data
df = pd.read_csv("data/final/merged_county_data.csv")
y = df["Total_ED"].astype(float)
X = df.drop(columns=["County","Region","Total_ED"]).apply(pd.to_numeric, errors="coerce").fillna(0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# --- Ridge model (parametric)
ridge = Pipeline([
    ("scale", StandardScaler()),
    ("model", RidgeCV(alphas=np.logspace(-3, 3, 50)))
])
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)

print("R:", r2_score(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))

# --- Coefficient importance
coefs = pd.Series(ridge.named_steps["model"].coef_, index=X.columns)
coefs.sort_values(key=np.abs, ascending=False).head(20).plot(kind="barh")
plt.title("Top Ridge Coefficients")
plt.show()

# --- Optional Polynomial Ridge (for smooth nonlinearities)
poly_ridge = Pipeline([
    ("poly", PolynomialFeatures(degree=4, include_bias=False)),
    ("scale", StandardScaler()),
    ("model", RidgeCV(alphas=np.logspace(-3, 3, 50)))
])
poly_ridge.fit(X_train, y_train)
print("Poly Ridge R:", poly_ridge.score(X_test, y_test))

# --- Optional Random Forest for comparison
rf = RandomForestRegressor(n_estimators=200, random_state=42)
rf.fit(X_train, y_train)
print("RF R:", rf.score(X_test, y_test))
```
